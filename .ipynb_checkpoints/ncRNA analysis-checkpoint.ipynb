```python
import pandas as pd
from Bio import SeqIO

# Step 1: Load BLAST results
cols = ['qseqid','sseqid','pident','length','mismatch','gapopen','qstart','qend','sstart','send','evalue','bitscore']
blast_df = pd.read_csv('/home/vaibhavi/Downloads/spinal_cord/blast_results.txt', sep='\t', names=cols)

# Step 2: Compute transcript lengths from FASTA
fasta_path = "/home/vaibhavi/Downloads/spinal_cord/novel_transcripts.fa"
transcript_lengths = {record.id: len(record.seq) for record in SeqIO.parse(fasta_path, "fasta")}

# Step 3: Calculate query coverage
blast_df['qcov'] = (blast_df['qend'] - blast_df['qstart'] + 1) / blast_df['qseqid'].map(transcript_lengths)

# Step 4: Filter known (high identity + high coverage)
identity_threshold = 80
coverage_threshold = 0.8
known = blast_df[(blast_df['pident'] >= identity_threshold) & (blast_df['qcov'] >= coverage_threshold)]
known_transcripts = set(known['qseqid'])

# Step 5: Load CPC2 noncoding transcripts
cpc2 = pd.read_csv('/home/vaibhavi/Downloads/spinal_cord/cpc2_results.txt.txt', sep='\t')
noncoding_transcripts = set(cpc2[cpc2['label'] == 'noncoding']['#ID'])

# Step 6: Get novel lncRNAs = noncoding - known
novel_lncrnas = [t for t in noncoding_transcripts if t not in known_transcripts]

# Step 7: Output result
print(f"✅ Number of novel lncRNAs identified: {len(novel_lncrnas)}")

# (Optional) Save IDs to a file
with open("/home/vaibhavi/Downloads/spinal_cord/novel_lncRNAs_ids.txt", "w") as f:
    for tid in novel_lncrnas:
        f.write(tid + "\n")
```

```python
import pandas as pd
from Bio import SeqIO
import re

# ==== Step 1: Load BLAST Results ====
blast_cols = ['qseqid','sseqid','pident','length','mismatch','gapopen','qstart','qend','sstart','send','evalue','bitscore']
blast_df = pd.read_csv('/home/vaibhavi/Downloads/spinal_cord/blast_results.txt', sep='\t', names=blast_cols)

# ==== Step 2: Load transcript lengths from FASTA ====
fasta_path = "/home/vaibhavi/Downloads/spinal_cord/novel_transcripts.fa"
transcript_lengths = {record.id: len(record.seq) for record in SeqIO.parse(fasta_path, "fasta")}

# ==== Step 3: Compute query coverage ====
blast_df['qcov'] = (blast_df['qend'] - blast_df['qstart'] + 1) / blast_df['qseqid'].map(transcript_lengths)

# ==== Step 4: Filter known transcripts (based on identity & coverage thresholds) ====
identity_threshold = 80
coverage_threshold = 0.8
known = blast_df[(blast_df['pident'] >= identity_threshold) & (blast_df['qcov'] >= coverage_threshold)]
known_transcripts = set(known['qseqid'])

# ==== Step 5: Load CPC2 results ====
cpc2 = pd.read_csv('/home/vaibhavi/Downloads/spinal_cord/cpc2_results.txt.txt', sep='\t')
noncoding_transcripts = set(cpc2[cpc2['label'] == 'noncoding']['#ID'])

# ==== Step 6: Identify novel lncRNAs ====
novel_lncrnas = [t for t in noncoding_transcripts if t not in known_transcripts]

# ==== Step 7: Parse GTF to get gene_id and gene_name ====
gtf_path = "/home/vaibhavi/Downloads/spinal_cord/merged_all_samples.gtf"
mapping = {}

with open(gtf_path) as gtf_file:
    for line in gtf_file:
        if line.startswith('#'):
            continue
        fields = line.strip().split('\t')
        if fields[2] != "transcript":
            continue
        attr_field = fields[8]
        try:
            transcript_id = re.search('transcript_id "([^"]+)"', attr_field).group(1)
            gene_id = re.search('gene_id "([^"]+)"', attr_field).group(1)
            gene_name_match = re.search('gene_name "([^"]+)"', attr_field)
            gene_name = gene_name_match.group(1) if gene_name_match else "-"
            mapping[transcript_id] = {'gene_id': gene_id, 'gene_name': gene_name}
        except Exception:
            continue

# ==== Step 8: Collect novel lncRNA info ====
novel_lncrna_info = []
for transcript in novel_lncrnas:
    gene_info = mapping.get(transcript, {'gene_id': 'NA', 'gene_name': 'NA'})
    novel_lncrna_info.append({
        'transcript_id': transcript,
        'gene_id': gene_info['gene_id'],
        'gene_name': gene_info['gene_name']
    })

# ==== Step 9: Save to CSV ====
novel_df = pd.DataFrame(novel_lncrna_info)
output_path = "/home/vaibhavi/Downloads/spinal_cord/novel_lncRNAs_with_gene_info.csv"
novel_df.to_csv(output_path, index=False)

print(f"✅ Total novel lncRNAs with gene info: {len(novel_df)}")
print(f"📄 Output saved to: {output_path}")
```

```python
import pandas as pd
import re

# === Step 1: Load your novel lncRNA IDs ===
with open("/home/vaibhavi/Downloads/spinal_cord/novel_lncRNAs_ids.txt") as f:
    novel_lncrnas = set(line.strip() for line in f)

# === Step 2: Extract annotation info from merged GTF ===
gtf_file = "/home/vaibhavi/Downloads/spinal_cord/merged_all_samples.gtf"

transcript_info = []

with open(gtf_file, "r") as gtf:
    for line in gtf:
        if line.startswith("#"):
            continue
        fields = line.strip().split('\t')
        if len(fields) < 9:
            continue
        feature = fields[2]
        if feature != "transcript":
            continue
        
        chrom, source, feat, start, end, score, strand, frame, attributes = fields
        
        # Extract transcript_id, gene_id, and gene_name
        tid_match = re.search('transcript_id "([^"]+)"', attributes)
        gid_match = re.search('gene_id "([^"]+)"', attributes)
        gname_match = re.search('gene_name "([^"]+)"', attributes)
        
        if tid_match:
            tid = tid_match.group(1)
            if tid in novel_lncrnas:
                transcript_info.append({
                    "transcript_id": tid,
                    "gene_id": gid_match.group(1) if gid_match else "NA",
                    "gene_name": gname_match.group(1) if gname_match else "NA",
                    "chromosome": chrom,
                    "start": int(start),
                    "end": int(end),
                    "strand": strand
                })

# Convert to DataFrame
annotation_df = pd.DataFrame(transcript_info)

# === Step 3: Load TPM/FPKM expression data (Optional) ===
# Make sure your TPM matrix has 'transcript_id' as the first column
tpm_file = "/home/vaibhavi/Downloads/spinal_cord/tpm_matrix.csv"  # Replace with your actual file
try:
    tpm_df = pd.read_csv(tpm_file)

    # Merge annotation with expression
    final_df = annotation_df.merge(tpm_df, on="transcript_id", how="left")

    # Add average TPM column
    tpm_columns = [col for col in final_df.columns if "TPM" in col]
    if tpm_columns:
        final_df["mean_TPM"] = final_df[tpm_columns].mean(axis=1)

except FileNotFoundError:
    print("TPM matrix not found — skipping expression merge.")
    final_df = annotation_df

# === Step 4: Save the output ===
output_file = "/home/vaibhavi/Downloads/spinal_cord/novel_lncRNAs_full_annotation.csv"
final_df.to_csv(output_file, index=False)
print(f"Full annotation saved to: {output_file}")
```

# correlation of mrna and lncrna using pearson value

```python
# Load required libraries
library(dplyr)

# Step 1: Load expression matrices
lncrna_counts <- read.delim("/home/vaibhavi/Downloads/spinal_cord/lncrna_count..csv", header = TRUE, sep = "\t", stringsAsFactors = FALSE)
rownames(lncrna_counts) <- lncrna_counts$lncRNA_ID
lncrna_counts <- lncrna_counts[, -1]  # Remove lncRNA_ID column

mrna_counts <- read.csv("/home/vaibhavi/Downloads/spinal_cord/filtered_mrna_counts.csv", row.names = 1, check.names = FALSE)

# Step 2: Load overlap file
overlaps <- read.csv("/home/vaibhavi/Downloads/spinal_cord/lncRNA_mRNA_overlaps_full.csv", stringsAsFactors = FALSE)

# Step 3: Rename columns if needed
colnames(overlaps)[colnames(overlaps) == "lncRNA_id"] <- "lncRNA_ID"
colnames(overlaps)[colnames(overlaps) == "mRNA_gene_id"] <- "mRNA_ID"

# Step 4: Drop rows with missing IDs
overlaps <- overlaps %>% filter(!is.na(lncRNA_ID), !is.na(mRNA_ID))

# Step 5: Get unique lncRNA–mRNA pairs
lnc_mrna_pairs <- overlaps[, c("lncRNA_ID", "mRNA_ID")] %>% distinct()

# Step 6: Initialize results dataframe
results <- data.frame(
  lncRNA = character(),
  mRNA = character(),
  correlation = numeric(),
  pvalue = numeric(),
  stringsAsFactors = FALSE
)

# Step 7: Correlation analysis
for (i in 1:nrow(lnc_mrna_pairs)) {
  lnc_id <- lnc_mrna_pairs$lncRNA_ID[i]
  mrna_id <- lnc_mrna_pairs$mRNA_ID[i]
  
  if (lnc_id %in% rownames(lncrna_counts) && mrna_id %in% rownames(mrna_counts)) {
    lnc_expr <- as.numeric(lncrna_counts[lnc_id, ])
    mrna_expr <- as.numeric(mrna_counts[mrna_id, ])
    
    # Check if both have non-zero variance
    if (sd(lnc_expr) != 0 && sd(mrna_expr) != 0) {
      cor_test <- cor.test(lnc_expr, mrna_expr, method = "pearson")
      results <- rbind(results, data.frame(
        lncRNA = lnc_id,
        mRNA = mrna_id,
        correlation = cor_test$estimate,
        pvalue = cor_test$p.value
      ))
    }
  }
}

# Step 8: Save correlation results
write.csv(results, "lncRNA_mRNA_correlation_results.csv", row.names = FALSE)

# Step 9: Print summary
cat("\nCorrelation analysis completed.")
cat("\nNumber of correlated pairs:", nrow(results), "\n")

```

```python
import pandas as pd
import re

gtf_path = "/home/vaibhavi/Downloads/spinal_cord/known_mRNA.gtf"
output_csv_path = "/home/vaibhavi/Downloads/spinal_cord/known_mRNA_gene_annotation.csv"

# Function to parse GTF attributes column
def parse_gtf_attributes(attr_string):
    attr_dict = {}
    for attr in attr_string.strip().split(';'):
        if attr.strip():
            try:
                key, value = attr.strip().split(' ', 1)
                attr_dict[key] = value.strip('"')
            except ValueError:
                continue
    return attr_dict

records = []
with open(gtf_path, 'r') as f:
    for line in f:
        if line.startswith('#'):
            continue
        fields = line.strip().split('\t')
        if len(fields) < 9:
            continue
        chrom, source, feature, start, end, score, strand, frame, attributes = fields
        # Choose feature 'gene' or 'transcript'; 'gene' preferred for gene-level annotation
        if feature == 'gene':
            attr_dict = parse_gtf_attributes(attributes)
            gene_id = attr_dict.get('gene_id', '')
            if gene_id:
                records.append({
                    'chromosome': chrom,
                    'start': int(start),
                    'end': int(end),
                    'gene_id': gene_id
                })

# Create DataFrame and save
df = pd.DataFrame(records)
df.to_csv(output_csv_path, index=False)

print(f"Gene annotation CSV saved to: {output_csv_path}")
print(f"Total genes extracted: {len(df)}")
```

# Overlapping lncnrna with mrna gene id

```python
import pandas as pd
import pyranges as pr

# Load novel lncRNA annotation
lncrna_df = pd.read_csv("/home/vaibhavi/Downloads/spinal_cord/novel_lncRNAs_full_annotation.csv")

# Load known mRNA annotation
pc_genes_df = pd.read_csv("/home/vaibhavi/Downloads/spinal_cord/known_mRNA_gene_annotation.csv")

# Standardize chromosome names (remove "chr" prefix if present)
lncrna_df["chromosome"] = lncrna_df["chromosome"].astype(str).str.replace("^chr", "", regex=True)
pc_genes_df["chromosome"] = pc_genes_df["chromosome"].astype(str).str.replace("^chr", "", regex=True)

# Rename columns to match PyRanges format
lncrna_df = lncrna_df.rename(columns={"chromosome": "Chromosome", "start": "Start", "end": "End"})
pc_genes_df = pc_genes_df.rename(columns={"chromosome": "Chromosome", "start": "Start", "end": "End"})

# Create PyRanges objects
lncrna_gr = pr.PyRanges(lncrna_df)
pc_genes_gr = pr.PyRanges(pc_genes_df)

# Extend lncRNA coordinates ±100 kb
lncrna_gr_window = lncrna_gr.slack(100_000)

# Find overlaps
overlaps = lncrna_gr_window.join(pc_genes_gr)
overlaps_df = overlaps.df

# Check if any overlaps were found
if overlaps_df.empty:
    print("❌ No overlaps found between lncRNAs and mRNAs.")
else:
    # Show all available columns (for debugging if needed)
    print("🧪 Columns in overlaps_df:", overlaps_df.columns.tolist())

    # Determine columns for lncRNA coordinates
    lnc_cols = ["Chromosome", "Start", "End"]

    # Determine mRNA coordinate columns with "_b" suffix
    mrna_cols = [col for col in ["Chromosome_b", "Start_b", "End_b"] if col in overlaps_df.columns]

    # Determine ID columns
    lncRNA_id_col = "transcript_id" if "transcript_id" in overlaps_df.columns else "gene_id"
    mrna_gene_id_col = None
    for col in overlaps_df.columns:
        if col.startswith("gene_id") and col.endswith("_b"):
            mrna_gene_id_col = col
            break
    if not mrna_gene_id_col:
        for col in overlaps_df.columns:
            if col == "gene_id":
                mrna_gene_id_col = col
                break

    if mrna_gene_id_col is None:
        raise ValueError("❌ Could not find mRNA gene_id column after join.")

    # Final columns to export
    columns_to_export = lnc_cols + mrna_cols + [lncRNA_id_col, mrna_gene_id_col]

    # Copy selected columns
    output_df = overlaps_df[columns_to_export].copy()

    # Rename for clarity
    rename_dict = {
        "Chromosome": "lncRNA_chr",
        "Start": "lncRNA_start",
        "End": "lncRNA_end",
        lncRNA_id_col: "lncRNA_id",
        mrna_gene_id_col: "mRNA_gene_id"
    }
    if len(mrna_cols) == 3:
        rename_dict.update({
            "Chromosome_b": "mRNA_chr",
            "Start_b": "mRNA_start",
            "End_b": "mRNA_end"
        })

    output_df.rename(columns=rename_dict, inplace=True)
    output_df = output_df.drop_duplicates()

    # Save to CSV
    output_path = "/home/vaibhavi/Downloads/spinal_cord/lncRNA_mRNA_overlaps_full.csv"
    output_df.to_csv(output_path, index=False)

    print(f"✅ Found and saved {len(output_df)} overlapping lncRNA–mRNA pairs with coordinates.")
    print(f"📁 Output: {output_path}")
```

```python
import pandas as pd

# Load overlaps file (lncRNA-mRNA pairs)
overlap_df = pd.read_csv("/home/vaibhavi/Downloads/spinal_cord/lncRNA_mRNA_overlaps_full.csv")

# Load logFC and p-values for transcripts (could include both lncRNAs and mRNAs)
logfc_df = pd.read_csv("/home/vaibhavi/Downloads/spinal_cord/logfc_with_pvalues.csv")

# Merge logFC info for lncRNAs
merged = pd.merge(overlap_df, logfc_df, left_on="lncRNA_id", right_on="transcript_id", how="left")
merged = merged.rename(columns={"logFC": "lncRNA_logFC", "pvalue": "lncRNA_pvalue"})
merged.drop(columns=["transcript_id"], inplace=True)

# Merge logFC info for mRNAs
merged = pd.merge(merged, logfc_df, left_on="mRNA_gene_id", right_on="transcript_id", how="left")
merged = merged.rename(columns={"logFC": "mRNA_logFC", "pvalue": "mRNA_pvalue"})
merged.drop(columns=["transcript_id"], inplace=True)

# Define criteria
def is_sig(logfc, pval):
    return abs(logfc) >= 1 and pval < 0.05

filtered = merged[
    merged["lncRNA_logFC"].notnull() &
    merged["mRNA_logFC"].notnull() &
    merged.apply(lambda row: is_sig(row["lncRNA_logFC"], row["lncRNA_pvalue"]) or 
                             is_sig(row["mRNA_logFC"], row["mRNA_pvalue"]), axis=1)
]

output_path = "/home/vaibhavi/Downloads/spinal_cord/lncRNA_mRNA_overlap_with_logfc.csv"
filtered.to_csv(output_path, index=False)
print(f"✅ Annotated overlap saved: {output_path}")

import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(filtered["lncRNA_logFC"], filtered["mRNA_logFC"], alpha=0.7)
plt.xlabel("lncRNA logFC")
plt.ylabel("mRNA logFC")
plt.title("lncRNA vs mRNA logFC (overlapping pairs)")
plt.axhline(0, linestyle='--', color='gray')
plt.axvline(0, linestyle='--', color='gray')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Step 1: Load overlap data (lncRNA ↔ mRNA mapping)
overlap_df = pd.read_csv("/home/vaibhavi/Downloads/spinal_cord/lncRNA_mRNA_overlaps_full.csv")

# Step 2: Load logFC expression data
logfc_df = pd.read_csv("/home/vaibhavi/Downloads/spinal_cord/logfc_with_pvalues.csv")

# Step 3: Merge logFC data with overlap info using lncRNA ID
# 'lncRNA_id' in overlap_df corresponds to 'transcript_id' in logfc_df
merged_df = pd.merge(logfc_df, 
                     overlap_df[['lncRNA_id', 'mRNA_gene_id']], 
                     left_on='transcript_id', 
                     right_on='lncRNA_id', 
                     how='left')

# Optional: Drop the duplicated lncRNA_id column
merged_df.drop(columns=['lncRNA_id'], inplace=True)

# Step 4: Save the merged result
merged_df.to_csv("/home/vaibhavi/Downloads/spinal_cord/logfc_with_gene_ids.csv", index=False)

print("✅ Merged file saved: logfc_with_gene_ids.csv")
```

```python
# Install required packages if not already installed
if (!requireNamespace("BiocManager", quietly = TRUE))
  install.packages("BiocManager")

BiocManager::install(c("clusterProfiler", "org.Hs.eg.db", "AnnotationDbi", "enrichplot", "ggplot2", "dplyr"))

# Load libraries
library(clusterProfiler)
library(org.Hs.eg.db)
library(AnnotationDbi)
library(enrichplot)
library(ggplot2)
library(dplyr)

# Step 1: Load the dataset
data <- read.csv("/home/vaibhavi/Downloads/spinal_cord/logfc_with_gene_ids.csv")

# Step 2: Clean gene IDs (remove version numbers)
data$mRNA_gene_id_clean <- sub("\\..*", "", data$mRNA_gene_id)
unique_genes <- unique(data$mRNA_gene_id_clean)

# Step 3: Convert Ensembl IDs to Entrez IDs
entrez_ids <- bitr(unique_genes,
                   fromType = "ENSEMBL",
                   toType = "ENTREZID",
                   OrgDb = org.Hs.eg.db)

# Step 4: Prepare gene list for enrichment
gene_list <- na.omit(entrez_ids$ENTREZID)

# Step 5: GO Enrichment (ALL: BP, MF, CC)
go_enrich <- enrichGO(gene          = gene_list,
                      OrgDb         = org.Hs.eg.db,
                      ont           = "ALL",
                      pAdjustMethod = "BH",
                      pvalueCutoff  = 0.05,
                      readable      = TRUE)

# Step 6: KEGG Enrichment
kegg_enrich <- enrichKEGG(gene         = gene_list,
                          organism     = 'hsa',
                          pvalueCutoff = 0.05)

# Step 7: Save Results
write.csv(as.data.frame(go_enrich), "/home/vaibhavi/Downloads/spinal_cord/GO_Enrichment_Results.csv", row.names = FALSE)
write.csv(as.data.frame(kegg_enrich), "/home/vaibhavi/Downloads/spinal_cord/KEGG_Enrichment_Results.csv", row.names = FALSE)

# Step 8: Plot GO Enrichment - Barplot
barplot(go_enrich, showCategory = 10, title = "Top 10 GO Terms") +
  theme(axis.text.x = element_text(size = 12, angle = 45, hjust = 1),
        axis.text.y = element_text(size = 12),
        plot.title = element_text(size = 16, face = "bold"),
        axis.title.x = element_text(size = 14),
        axis.title.y = element_text(size = 14))

# Step 9: Plot KEGG Enrichment - Barplot
barplot(kegg_enrich, showCategory = 10, title = "Top 10 KEGG Pathways") +
  theme(axis.text.x = element_text(size = 12, angle = 45, hjust = 1),
        axis.text.y = element_text(size = 12),
        plot.title = element_text(size = 16, face = "bold"),
        axis.title.x = element_text(size = 14),
        axis.title.y = element_text(size = 14))

# Step 10: GO Dotplot
dotplot(go_enrich, showCategory = 10, title = "GO Dotplot") +
  theme(axis.text.x = element_text(size = 12, angle = 45, hjust = 1),
        axis.text.y = element_text(size = 12),
        plot.title = element_text(size = 16, face = "bold"),
        axis.title.x = element_text(size = 14),
        axis.title.y = element_text(size = 14))

# Step 11: KEGG Dotplot
dotplot(kegg_enrich, showCategory = 10, title = "KEGG Dotplot") +
  theme(axis.text.x = element_text(size = 12, angle = 45, hjust = 1),
        axis.text.y = element_text(size = 12),
        plot.title = element_text(size = 16, face = "bold"),
        axis.title.x = element_text(size = 14),
        axis.title.y = element_text(size = 14))

# Optional: Enrichment Map
emapplot(go_enrich)
```
